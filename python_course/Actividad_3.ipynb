## Ejercicio 3

Define una función llamada is_romano() dentro de la clase Numero que a partir de una cadena de caracteres, devuelve True si esa cadena de caracteres corresponde con un número romano y falso en caso contrario. Después, modifica el método para que lance un error en el caso de que el valor que nos pasan por parámetro no se corresponde con el patrón de un número romano.

```python
import re

class Numero():
    def __init__(self, numero) -> None:
        self.normal = numero
        self.romano = self.int_to_roman(numero)

    def imprime(self):
        print(self.normal, self.romano)

    def suma_romano(self, roman):
        if self.is_romano(roman):
            roman2int = self.roman_to_int(roman)
            self.normal = self.normal + roman2int
            self.romano = self.int_to_roman(self.normal)
        else: 
            raise Exception("El numero entregado no es un numero romano valido")

    def int_to_roman(self, num):
        roman_numerals = {
            1: 'I', 4: 'IV', 5: 'V', 9: 'IX', 10: 'X',
            40: 'XL', 50: 'L', 90: 'XC', 100: 'C',
            400: 'CD', 500: 'D', 900: 'CM', 1000: 'M'
        }
        result = ''
        for value, numeral in sorted(roman_numerals.items(), reverse=True):
            while num >= value:
                result += numeral
                num -= value
        return result
    
    def roman_to_int(self, roman):
        roman_numerals = {
            'I': 1, 'V': 5, 'X': 10, 'L': 50,
            'C': 100, 'D': 500, 'M': 1000
        }
        result = 0
        prev_value = 0
        for numeral in reversed(roman):
            value = roman_numerals[numeral]
            if value < prev_value:
                result -= value
            else:
                result += value
            prev_value = value
        return result
    
    def is_romano(self, roman):
        try:
            pattern = r'^M{0,3}(CM|CD|D?C{0,3})(XC|XL|L?X{0,3})(IX|IV|V?I{0,3})$'
            return re.match(pattern, roman) is not None
        except TypeError:
            pass
```

**NOTA:** el mensaje de error ha de seguir el siguiente formato --> print('Ha fallado el número', 5). 

```python
class MejorNumero(Numero):
    def resta(self, roman):
        if self.is_romano(roman):
            roman2int = self.roman_to_int(roman)
            self.normal = self.normal - roman2int
            self.romano = self.int_to_roman(self.normal)
        else: 
            raise Exception("El numero entregado no es un numero romano valido")
        
    def multiplica(self, roman):
        if self.is_romano(roman):
            roman2int = self.roman_to_int(roman)
            self.normal = self.normal * roman2int
            self.romano = self.int_to_roman(self.normal)
        else: 
            raise Exception("El numero entregado no es un numero romano valido")
        
    def iterar(self, lista):
        for numero in lista:
            if self.is_romano(numero):
                self.suma_romano(numero)
            else: 
                print('Ha fallado el número', numero)
```
